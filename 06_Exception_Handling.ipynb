## Project 1: Purchase Bill Analyzer

You had saved the items and their price details in a text file, which you purchased yesterday from a nearby super market.

You need to know:
1. How many items did you purchase?
2. How many items are free?
3. What is the total amount you had to pay?
4. What is the discount amount?
5. What is the final amount did you pay after the discount?

**Note:**
- Data is stored in a text file (e.g., Purchase-1.txt).
- Each line contains one item's details (item name and price separated by a space).
- You have to enter the file name during run time.
- Include necessary code to handle the possible exceptions.

**Sample Output 1:**

````python
%%writefile Purchase-1.txt
Chocolate 50
Biscuit 35
Icecream 50

Discount 5
````

````python
%%writefile Purchase-2.txt
Chocolate 50
Biscuit 35
Icecream 50
Rice 100
Chicken 250

Perfume Free
Soup Free

Discount 80
````

````python
def analyze_purchase(filename):
    try:
        with open(filename, 'r') as f:
            lines = f.readlines()
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
        return
    except Exception as e:
        print(f"Error reading file: {e}")
        return

    items_purchased = 0
    free_items = 0
    amount = 0
    discount = 0
    in_free_section = False

    for line in lines:
        line = line.strip()
        if not line:
            in_free_section = True
            continue

        parts = line.split()
        if not parts:
            continue

        if parts[0].lower() == 'discount':
            try:
                discount = int(parts[1])
            except (IndexError, ValueError):
                print("Error: Invalid discount value.")
            continue

        if not in_free_section:
            items_purchased += 1
            try:
                amount += int(parts[-1])
            except ValueError:
                print(f"Error: Invalid price for item '{parts[0]}'.")
        else:
            free_items += 1

    final_amount = amount - discount

    print(f"No of items purchased: {items_purchased}")
    print(f"No of free items: {free_items}")
    print(f"Amount to pay: {amount}")
    print(f"Discount given: {discount}")
    print(f"Final amount paid: {final_amount}")

filename = input("Enter the file name: ")
if not filename.endswith('.txt'):
    filename += '.txt'
analyze_purchase(filename)
````

### 1. Division with Exception Handling
Write a program to accept two numbers from the user and perform division. If any exception occurs, print an error message or else print the result.

````python
try:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    result = a / b
    print(f"Result: {result}")
except ValueError:
    print("Error: Please enter valid integers.")
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except Exception as e:
    print(f"Error: {e}")
````

### 2. Prime Check with Exception Handling
Write a program to accept a number from the user and check whether it's prime or not. If user enters anything other than number, handle the exception and print an error message.

````python
try:
    num = int(input("Enter a number: "))
    if num < 2:
        print(f"{num} is not prime")
    else:
        is_prime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print(f"{num} is prime")
        else:
            print(f"{num} is not prime")
except ValueError:
    print("Error: Please enter a valid integer.")
````

### 3. File Open with Exception Handling
Write a program to accept the file name to be opened from the user. If file exists, print the contents of the file in title case or else handle the exception and print an error message.

````python
filename = input("Enter file name: ")
try:
    with open(filename, 'r') as f:
        content = f.read()
    print(content.title())
except FileNotFoundError:
    print(f"Error: File '{filename}' not found.")
except Exception as e:
    print(f"Error: {e}")
````

### 4. List Index with Exception Handling
Declare a list with 10 integers and ask the user to enter an index. Check whether the number in that index is positive or negative number. If any invalid index is entered, handle the exception and print an error message.

````python
my_list = [10, -5, 23, -8, 0, 45, -12, 7, 99, -3]

try:
    index = int(input(f"Enter an index (0-{len(my_list)-1}): "))
    value = my_list[index]
    if value > 0:
        print(f"{value} is positive")
    elif value < 0:
        print(f"{value} is negative")
    else:
        print(f"{value} is zero")
except IndexError:
    print(f"Error: Index out of range. Please enter between 0 and {len(my_list)-1}.")
except ValueError:
    print("Error: Please enter a valid integer index.")
except Exception as e:
    print(f"Error: {e}")
````
