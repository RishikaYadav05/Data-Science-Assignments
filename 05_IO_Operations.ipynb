## Project 1: Secret Message

Your friend has sent you a text file containing n lines. He sent a secret message with it, which tells you the place and time where you have to go and meet him.

**Hints:**
1. The number of lines in the file tells you the meeting time.
   - 1 <= number of lines <= 24
   - If the number of lines exceeds 12, convert to 12-hour format.
     - e.g., 15 lines → 3 PM, 10 lines → 10 AM
2. The word appearing the maximum number of times tells you the meeting place (street name).
   - e.g., if "Oxford" appears most, meeting place is "Oxford Street".

**Sample Input 1 (sample.txt):**
```
Cricket, a bat-and-ball park game played between two teams of eleven park players on a field at the park center of which is a 20-metre (22-yard) pitch with a wicket at each end, each park comprising two bails balanced on three stumps. The batting park scores runs by striking the ball bowled at the park wicket with the park bat, while the bowling and park fielding side tries to prevent this and dismiss each park player (so they are "out"). Means of park include being bowled, when the ball hits the park and dislodges the bails, and by the fielding side park the ball after it is hit by the bat, but before it hits the park. When ten park have been dismissed, the innings ends and the teams park roles.
```
**Sample Output 1:**
```
Meeting time: 9 AM
Meeting place: Park Street
```

---

First, create the sample.txt file with the cricket paragraph (9 lines).

````python
%%writefile sample.txt
Cricket, a bat-and-ball park game played between two teams of eleven park players on a field at the park center of which is a 20-metre (22-yard) pitch with a wicket at each end, each park comprising two bails balanced on three stumps.
The batting park scores runs by striking the ball bowled at the park wicket with the park bat, while the bowling and park fielding side tries to prevent this and dismiss each park player (so they are "out").
Means of park include being bowled, when the ball hits the park and dislodges the bails, and by the fielding side park the ball after it is hit by the bat, but before it hits the park.
When ten park have been dismissed, the innings ends and the teams park roles.
Cricket is a popular sport in many countries.
It is played with a bat and ball on a rectangular pitch.
The game originated in England and spread across the world.
International matches can last up to five days.
The sport has millions of fans globally.
````

````python
def secret_message(filename):
    with open(filename, 'r') as f:
        content = f.read()
        lines = content.strip().split('\n')

    num_lines = len(lines)

    if num_lines > 12:
        hour = num_lines - 12
        meridian = "PM"
    else:
        hour = num_lines
        meridian = "AM"

    words = content.split()
    cleaned = [w.strip('.,!?();:\"\'').lower() for w in words]
    freq = {}
    for w in cleaned:
        if w:
            freq[w] = freq.get(w, 0) + 1

    max_word = max(freq, key=freq.get)

    print(f"Meeting time: {hour} {meridian}")
    print(f"Meeting place: {max_word.capitalize()} Street")

secret_message("sample.txt")
````

### 2. Read First n Lines
Write a program to read first n lines from a txt file. Get n as user input.

````python
n = int(input("Enter number of lines to read: "))
with open("demo.txt", 'r') as f:
    for i in range(n):
        line = f.readline()
        if line:
            print(line, end='')
````

### 3. Append User Input to File
Write a program to accept input from user and append it to a txt file.

````python
text = input("Enter text to append: ")
with open("demo.txt", 'a') as f:
    f.write(text + '\n')
print("Appended successfully!")
````

### 4. Read Lines into a List
Write a program to read contents from a txt file line by line and store each line into a list.

````python
lines_list = []
with open("demo.txt", 'r') as f:
    for line in f:
        lines_list.append(line.rstrip('\n'))
print(lines_list)
````

### 5. Find the Longest Word
Write a program to find the longest word from the txt file contents, assuming that the file will have only one longest word in it.

````python
with open("demo.txt", 'r') as f:
    words = f.read().split()

longest = max(words, key=len)
print(f"Longest word: {longest}")
````

### 6. Count Word Frequency
Write a program to count the frequency of a user entered word in a txt file.

````python
search_word = input("Enter a word to count: ").lower()

with open("demo.txt", 'r') as f:
    content = f.read().lower()

words = content.split()
count = words.count(search_word)
print(f"'{search_word}' appears {count} time(s)")
````
