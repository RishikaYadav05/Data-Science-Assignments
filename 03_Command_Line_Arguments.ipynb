## Project 1: Happiness Calculator

Through command line arguments three strings separated by space are given to you. Each string is a series of numbers separated by hyphen(-). You like all the numbers in string 1 and dislike all the numbers in string 2. Third string contains the numbers given to you. Your initial happiness is 0. When you encounter a number which is present in string 1, add 1 to your happiness, if it is present in string 2, add -1 to your happiness. Otherwise, your happiness does not change. Output your final happiness at the end.

**Sample Input 1:**
```
3-1
5-7
1-5-3-8
```
**Sample Output 1:** `1`

**Explanation:**
- Numbers in string 1: 3, 1
- Numbers in string 2: 5, 7
- Numbers given to you: 1, 5, 3, 8
- You gain 1 unit of happiness for numbers 1 and 3 which are in string 1. Total happiness: 2
- You lose 1 unit of happiness for number 5 which is in string 2. Total happiness: 1
- 8 is not present in either string, so happiness does not change
- Final happiness: 1

**Sample Input 2:**
```
60-77-34-5-2
44-11-99-3
77-15-13-2-34-3
```
**Sample Output 2:** `2`

**Explanation:**
- You gain 1 unit for 77, 2, 34 (in string 1). Happiness: 3
- You lose 1 unit for 3 (in string 2). Happiness: 2
- 15, 13 are not in either string. No change
- Final happiness: 2

````python
import sys

if len(sys.argv) < 4:
    print("Usage: python script.py <like-string> <dislike-string> <given-string>")
else:
    like_set = set(sys.argv[1].split('-'))
    dislike_set = set(sys.argv[2].split('-'))
    given_list = sys.argv[3].split('-')

    happiness = 0
    for num in given_list:
        if num in like_set:
            happiness += 1
        elif num in dislike_set:
            happiness -= 1

    print(happiness)
````

### 1. Sum of Two Numbers
Write a program to accept two numbers as command line arguments and display their sum.

````python
import sys

if len(sys.argv) >= 3:
    a = int(sys.argv[1])
    b = int(sys.argv[2])
    print(f"Sum: {a + b}")
else:
    print("Please provide two numbers as command line arguments")
````

### 2. Welcome Message with Filename
Write a program to accept a welcome message through command line arguments and display the file name along with the welcome message.

````python
import sys

if len(sys.argv) >= 2:
    welcome = ' '.join(sys.argv[1:])
    print(f"{welcome} - from {sys.argv[0]}")
else:
    print(f"No welcome message provided. Script: {sys.argv[0]}")
````

### 3. Sum of Prime Numbers
Write a program to accept 10 numbers through command line arguments and calculate the sum of prime numbers among them.

````python
import sys

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

if len(sys.argv) >= 11:
    numbers = [int(x) for x in sys.argv[1:11]]
    prime_sum = sum(num for num in numbers if is_prime(num))
    print(f"Sum of prime numbers: {prime_sum}")
else:
    print("Please provide exactly 10 numbers as command line arguments")
````
